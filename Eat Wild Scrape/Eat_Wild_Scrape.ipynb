```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```python
page = requests.get('http://eatwild.com/products/washington.html').text
soup = BeautifulSoup(page, 'lxml')
```

```python
pageList = page.split()
```

```python
count = 0
for element in pageList:
#     print(element, count)
    count += 1
```

```python
num = 1
for element in pageList:
    if 'href="mailto:' in element:
        print(element.split(':')[1].split('"')[0], num)
        num += 1
```

```python
with open('emails.csv', 'w', newline='') as f:

    csv_file = csv.writer(f)
    csv_file.writerow(['Name', 'Location', 'Email', 'Phone'])
    num = 1
    for element in pageList:
        if 'href="mailto:' in element:
            email = element.split(':')[1].split('"')[0]
            csv_file.writerow(['unknown', 'unknown', email, 'unknown'])
            num += 1

```
