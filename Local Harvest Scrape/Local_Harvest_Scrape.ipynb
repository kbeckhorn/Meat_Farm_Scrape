```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```python
page = requests.get('https://www.localharvest.org/search.jsp?jmp&scale=8&lat=47.202698&lon=-123.09886&ty=0').text
soup = BeautifulSoup(page, 'lxml')
```

```python
sections = soup.find_all(class_ = 'membercell')

with open('name_location_csv.csv', 'w', newline='') as f:
    csv_file = csv.writer(f)
    csv_file.writerow(['Name', 'Location'])

    for element in sections:
        count = 1
        for info in element.find_all('a'):

            if count == 1:
                name = info.text
            if count == 2:
                location = info.text
                csv_file.writerow([name, location])
            count += 1

      
```
